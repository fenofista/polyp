```python
import os
import numpy as np
print(len(os.listdir('seg/images')))
print(len(os.listdir('seg/labels')))

print(len(os.listdir('od/images')))
print(len(os.listdir('od/labels')))
```

```python
image_dir = (os.listdir('od/images'))
label_dir = (os.listdir('od/labels'))
print(len(image_dir))
print(len(label_dir))
```

```python
from pathlib import Path

image_stems = {Path(f).stem for f in image_dir}
label_stems = {Path(f).stem for f in label_dir}

missing = image_stems - label_stems
print(len(missing))
print(missing)
```

```python
# 過濾掉非圖片/非label檔案再比對
valid_image_stems = {Path(f).stem for f in image_dir 
                     if Path(f).suffix in ('.jpg', '.png', '.jpeg')}
valid_label_stems = {Path(f).stem for f in label_dir 
                     if Path(f).suffix == '.txt'}

missing = valid_image_stems - valid_label_stems
extra_labels = valid_label_stems - valid_image_stems

print(f"有圖沒label: {len(missing)}")
print(f"有label沒圖: {len(extra_labels)}")
print(missing)
```
