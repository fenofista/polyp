# Update Paths in train.txt / val.txt

將 `train.txt` 和 `val.txt` 裡的路徑前綴換成 server 上的路徑。

```python
from pathlib import Path

# ── 設定 ──────────────────────────────────────────────────
# 要取代的舊路徑前綴（本機）
OLD_PREFIX = "/Users/tony.tu/Desktop/戴承智慧/polyp/yolo_data/od/"

version = "v3"  # ← 改這裡
# 換成的新路徑前綴（server 上的路徑）
NEW_PREFIX = f"/root/Desktop/polyp/yolo_data/od/"   # ← 改這裡

# 要處理的 txt 檔案（可以加更多）
TXT_FILES = [
    Path(f"od/{version}/train.txt"),
    Path(f"od/{version}/val.txt"),
    # Path("v1/seg/train.txt"),
    # Path("v1/seg/val.txt"),
]

# 預覽模式
DRY_RUN = True
```

```python
# 預覽
for txt_file in TXT_FILES:
    if not txt_file.exists():
        print(f"[SKIP] 找不到：{txt_file}")
        continue
    lines = txt_file.read_text().strip().splitlines()
    print(f"\n{txt_file}  ({len(lines)} 筆)")
    print(f"  舊：{lines[0]}")
    print(f"  新：{lines[0].replace(OLD_PREFIX, NEW_PREFIX, 1)}")
```

```python
DRY_RUN = False  # 確認後改為 False

if DRY_RUN:
    print("DRY RUN，不會修改任何檔案")
else:
    for txt_file in TXT_FILES:
        if not txt_file.exists():
            print(f"[SKIP] 找不到：{txt_file}")
            continue
        content = txt_file.read_text()
        new_content = content.replace(OLD_PREFIX, NEW_PREFIX)
        txt_file.write_text(new_content)
        n = content.count(OLD_PREFIX)
        print(f"{txt_file}: 取代 {n} 筆")
    print("完成")
```
